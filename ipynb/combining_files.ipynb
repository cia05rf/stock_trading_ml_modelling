# Combining files
This program is designed to take all the csv files in a folder and combine them into one file.

```python
#Import the libraries
import pandas as pd
import os
```

```python
#Setup the variables
src_fldr_pth = r'C:\\Users\\Robert\\Documents\\python_scripts\\stock_trading_ml_modelling\\historical_prices\\'
dst_fldr_pth = r'C:\\Users\\Robert\\Documents\\python_scripts\\stock_trading_ml_modelling\\historical_prices\\'
new_file_name = r'all_hist_prices_d'
file_ending = r'_d.csv'
group_name = r'daily_data'
```

```python
#Create a hdfs5 file
hf_store_name = dst_fldr_pth+new_file_name+".h5"
hf = pd.HDFStore(hf_store_name)
print(hf)
```

```python
#Get all the files in the folder and append to the hd5 file
count = 0
for file in os.listdir(src_fldr_pth):
    if file.endswith(file_ending):
        print("")
        count += 1
        print(count)
        print('READING IN: {}'.format(file))
        #Read in a csv file
        tmp_csv = pd.read_csv(src_fldr_pth+file)
        tmp_csv["Ticker"] = [str(x).upper() for x in tmp_csv["Ticker"]]
        tmp_csv["Index"] = tmp_csv["Index"].astype("int64")
        print("SHAPE: {}".format(tmp_csv.shape))
        #Append this data to the group
        try:
            tmp_csv.to_hdf(hf_store_name,key=group_name,append=True)
            print('ADDED TO {}'.format(hf_store_name))
        except Exception as e:
            print('ERROR READING TO FILE {}'.format(e))
hf.close()
```
